```python
from data_loader import load_data
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
train_data, train_labels, test_data = load_data()
```

```python
print(train_data.shape)
print(train_labels.shape)
print(test_data.shape)
```

```python
model = Sequential()
model.add(Dense(units=20, activation='relu', input_dim=12))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=4, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='sgd',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train_data, train_labels, epochs=20, batch_size=128)
```

```python
model.save('/home/raghuvansh/DL/SocialCops Land Classification Challenge/saved_network/model.h5')
```

```python
y_target = model.predict_classes(test_data) + 1
```

```python
import numpy as np
import pandas as pd
```

```python
output = np.zeros((len(y_target), 13))
df_test = pd.read_csv('/home/raghuvansh/DL/SocialCops Land Classification Challenge/data/land_test.csv')
output[:,:-1] = df_test.values
output[:,-1] = y_target
```

```python
df_output = pd.DataFrame(output)
```
